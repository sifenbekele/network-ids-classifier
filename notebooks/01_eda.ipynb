```python
df = df.replace([np.inf, -np.inf], np.nan)


rows_with_nan = df.isna().any(axis=1).sum()
print(f"Rows with NaN (including former inf): {rows_with_nan}")
print(f"Percentage of total: {rows_with_nan / len(df) * 100:.4f}%")


df = df.dropna()
print(f"New shape: {df.shape}")
```

```python
print((df['Flow Duration'] == -1).sum())
```

```python
df = df[df['Flow Duration'] != -1]
print(f"Shape after removing inf/NaN rows and -1 duration rows: {df.shape}")
```

```python
zero_var_cols = df.columns[df.nunique() <= 1]
print(f"Zero-variance columns ({len(zero_var_cols)}):")
print(zero_var_cols.tolist())
```

```python
df = df.drop(columns=zero_var_cols)
print(f"Shape after dropping zero-variance columns: {df.shape}")
```

```python
# Find columns that are exact duplicates of another column (by values)
duplicate_cols = []
cols = df.columns.tolist()
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        if df[cols[i]].equals(df[cols[j]]):
            duplicate_cols.append((cols[i], cols[j]))

print("Duplicate column pairs:")
print(duplicate_cols)
```

```python
cols_to_drop = [
    'Subflow Fwd Packets',
    'Subflow Bwd Packets',
    'Subflow Fwd Bytes',
    'Subflow Bwd Bytes',
    'Avg Fwd Segment Size',
    'Fwd PSH Flags',
    'Fwd Header Length.1',
    'ECE Flag Count'
]

df = df.drop(columns=cols_to_drop)
print(f"Shape after dropping duplicate columns: {df.shape}")
```

```python
df['Label_binary'] = (df['Label'] != 'BENIGN').astype(int)

print(df['Label_binary'].value_counts())
print(df[['Label', 'Label_binary']].head())
```

```python
# Separate features and labels
# Drop both label columns from X; keep the encoded binary label as y for now
X = df.drop(columns=['Label', 'Label_binary'])
y = df['Label_binary']

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"y distribution:\n{y.value_counts()}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train distribution:\n{y_train.value_counts(normalize=True)}")
print(f"y_test distribution:\n{y_test.value_counts(normalize=True)}")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_baseline = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_baseline.fit(X_train, y_train)

print("Baseline Random Forest trained.")
print(f"Train accuracy: {rf_baseline.score(X_train, y_train):.4f}")
print(f"Test accuracy: {rf_baseline.score(X_test, y_test):.4f}")
```

```python
import pandas as pd

importances = pd.Series(rf_baseline.feature_importances_, index=X_train.columns)
importances = importances.sort_values(ascending=False)
print(importances.head(20))
```

```python
import matplotlib.pyplot as plt

top_20 = importances.head(20)

plt.figure(figsize=(10, 8))
top_20.sort_values().plot(kind='barh')
plt.title('Top 20 Feature Importances - Baseline Random Forest')
plt.xlabel('Importance')
plt.tight_layout()
plt.show()
```

```python
# Compute correlation matrix on the training features only
corr_matrix = X_train.corr().abs()

# Get upper triangle of the matrix (avoid duplicate pairs and self-correlation)
upper_triangle = corr_matrix.where(
    pd.np.triu(pd.np.ones(corr_matrix.shape), k=1).astype(bool)
) if hasattr(pd, 'np') else corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

# Find pairs with correlation > 0.95
high_corr_pairs = []
for col in upper_triangle.columns:
    for row in upper_triangle.index:
        val = upper_triangle.loc[row, col]
        if pd.notna(val) and val > 0.95:
            high_corr_pairs.append((row, col, val))

high_corr_pairs = sorted(high_corr_pairs, key=lambda x: -x[2])
print(f"Found {len(high_corr_pairs)} feature pairs with correlation > 0.95:\n")
for f1, f2, v in high_corr_pairs:
    print(f"{f1}  <->  {f2}: {v:.3f}")
```

```python
# Build a set of features to drop, using importance as the tiebreaker
to_drop = set()

for f1, f2, v in high_corr_pairs:
    # If we've already decided to drop one of them, skip
    if f1 in to_drop or f2 in to_drop:
        continue
    # Drop whichever has lower importance
    if importances[f1] < importances[f2]:
        to_drop.add(f1)
    else:
        to_drop.add(f2)

print(f"Features to drop ({len(to_drop)}):")
print(sorted(to_drop))
```

```python
X_train_reduced = X_train.drop(columns=list(to_drop))
X_test_reduced = X_test.drop(columns=list(to_drop))

print(f"X_train_reduced shape: {X_train_reduced.shape}")
print(f"X_test_reduced shape: {X_test_reduced.shape}")
```

```python
rf_reduced = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_reduced.fit(X_train_reduced, y_train)

print(f"Reduced-feature train accuracy: {rf_reduced.score(X_train_reduced, y_train):.4f}")
print(f"Reduced-feature test accuracy: {rf_reduced.score(X_test_reduced, y_test):.4f}")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit on train only
X_train_scaled = scaler.fit_transform(X_train_reduced)

# Transform test using the same fitted scaler
X_test_scaled = scaler.transform(X_test_reduced)

# Convert back to DataFrames to keep column names
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train_reduced.columns, index=X_train_reduced.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test_reduced.columns, index=X_test_reduced.index)

print(X_train_scaled.describe().loc[['mean', 'std']])
```

```python
import joblib

joblib.dump(scaler, '../models/scaler.joblib')
print("Scaler saved to models/scaler.joblib")
```

```python
X_train_scaled.to_csv('../data/processed/X_train_scaled.csv', index=False)
X_test_scaled.to_csv('../data/processed/X_test_scaled.csv', index=False)

# Also save the unscaled reduced version, useful for tree-based models (RF, XGBoost) that don't need scaling
X_train_reduced.to_csv('../data/processed/X_train_reduced.csv', index=False)
X_test_reduced.to_csv('../data/processed/X_test_reduced.csv', index=False)

print("Saved reduced and scaled datasets to data/processed/")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

results = []

def evaluate_model(name, model, X_test, y_test):
    y_pred = model.predict(X_test)
    results.append({
        'model': name,
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred)
    })
    print(f"{name} — Accuracy: {results[-1]['accuracy']:.4f}, Precision: {results[-1]['precision']:.4f}, Recall: {results[-1]['recall']:.4f}, F1: {results[-1]['f1']:.4f}")
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_scaled, y_train)
evaluate_model('Logistic Regression', log_reg, X_test_scaled, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train_reduced, y_train)
evaluate_model('Decision Tree', dt, X_test_reduced, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train_reduced, y_train)
evaluate_model('Random Forest', rf, X_test_reduced, y_test)
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(random_state=42, n_jobs=-1, eval_metric='logloss')
xgb.fit(X_train_reduced, y_train)
evaluate_model('XGBoost', xgb, X_test_reduced, y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Logistic Regression with balanced class weights
log_reg_bal = LogisticRegression(max_iter=1000, random_state=42, class_weight='balanced')
log_reg_bal.fit(X_train_scaled, y_train)
evaluate_model('Logistic Regression (balanced)', log_reg_bal, X_test_scaled, y_test)

# Decision Tree with balanced class weights
dt_bal = DecisionTreeClassifier(random_state=42, class_weight='balanced')
dt_bal.fit(X_train_reduced, y_train)
evaluate_model('Decision Tree (balanced)', dt_bal, X_test_reduced, y_test)

# Random Forest with balanced class weights
rf_bal = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, class_weight='balanced')
rf_bal.fit(X_train_reduced, y_train)
evaluate_model('Random Forest (balanced)', rf_bal, X_test_reduced, y_test)

# XGBoost with scale_pos_weight (ratio of negative to positive class)
neg, pos = y_train.value_counts()[0], y_train.value_counts()[1]
scale_pos_weight = neg / pos
xgb_bal = XGBClassifier(random_state=42, n_jobs=-1, eval_metric='logloss', scale_pos_weight=scale_pos_weight)
xgb_bal.fit(X_train_reduced, y_train)
evaluate_model('XGBoost (balanced)', xgb_bal, X_test_reduced, y_test)
```
